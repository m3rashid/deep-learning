```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y,label='2x', color='#123fff', marker='o', linestyle='--')

plt.title('First basic graph', fontdict={'fontname': 'Droid Sans Mono', 'fontsize': '20'})
plt.xlabel('Some random label on x axis')
plt.ylabel('Some random label on y axis')

plt.xticks([1, 2, 3, 4, 5 ])
plt.yticks([2, 4, 6, 8, 10 ])

plt.legend()

plt.show()
```
